## Problem 2: Conceptual Questions

When we get high variance we get overfitting. Overfitting occurs when a model starts to memorize training data instead of learning to generalize a trend. Not letteing the algorithm run for long enough is a cause of overfitting.


    Overfitting is the case where the overall cost is really small, but the generalization of the model is unreliable. This is due to the model learning “too much” from the training data set.
        - Towards data science

If the line drawn over the data follows the data points exactly, we have a case of overfitting. 
To fix overfitting we need to use more training instances, reduce the number of features and/or use simpler models. To treat overfitting in regression models we can use ridge regression or lasso regression.

Sources:
 - https://en.wikipedia.org/wiki/Overfitting
 - https://towardsdatascience.com/what-are-overfitting-and-underfitting-in-machine-learning-a96b30864690

#### Ridge Regression 
Ridge regression can provide better long term predictions. It helps us reduce overfitting. The introduction of a small amount of bias gives us less variance. Using ridge regression we can solve for many paramaters using few samples. 
The disadvantage of using ridge regression is given the wrong lamda our model might get a high bias, giving us a poor model for predictions. We are dependent of using cross-validation (useually 10-fold cross validation) to find the right lamda, adding a additoonal step in finding the right model.

#### Lasso Regression
Lasso regression is also used to reduce overfitting. The diffrence between ridge regression and lasso regression is the penelty function. Where ridge regression squares the variables, Lasso regression takes the absolute value. Lasso regression can exclude useless variables from equations. This makes it a bit better than ridgde regression, reducing the variance in models containing useless variables.

Source: https://www.andreaperlato.com/theorypost/ridge-and-lasso-regression/

### 4.2 Defining the architecture

I will be using keras from tenserflow as my deep learing model. My model will have 2 hidden layers. I will use linear activation on my hidden layers and my output layer will use relu since we are getting posetive results. I will set my number of epochs and batch size trough trial and error. 

```python
import keras
print(keras.__version__)
```

```python
import tensorflow as tf
print(tf.__version__)
```

```python
from keras.models import Sequential
```

```python
from keras.layers import InputLayer, Dense 
```

### 4.3 Training the model

```python
input_neurons = X_train.shape[1]
input_neurons
```

```python
number_of_hidden_layers = 2
neuron_hidden_layer_1 = 10
neuron_hidden_layer_2 = 5
output_neurons = 1
```

```python
model = Sequential()
model.add(InputLayer(input_shape=(input_neurons)))
model.add(Dense(units=neuron_hidden_layer_1, activation='linear'))
model.add(Dense(units=neuron_hidden_layer_2, activation='linear'))
model.add(Dense(units=output_neurons, activation='relu'))
```

```python
model.summary()
```

```python
model.compile(
   loss = 'mse', 
   optimizer = 'Adam', 
   metrics = ['mean_absolute_error']
)
```

```python
from keras.callbacks import EarlyStopping
```

```python
history = model.fit(
   X_train,
   y_train,    
   batch_size = 100, 
   epochs = 1000, 
   verbose = 1, 
   validation_data = (X_test, y_test), 
   callbacks = [EarlyStopping(monitor = 'val_loss', patience = 100)]
)
```

```python
score = model.evaluate(X_test, y_test, verbose = 0) 
print('Test loss:', score[0]) 
print('Test accuracy:', score[1])
```

```python
predictions = model.predict(X_test)
predictions
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('mse')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```

```python
test_r2 = r2_score(y_test, predictions)

print(test_r2)
```
